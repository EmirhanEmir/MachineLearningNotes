```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
data1 = np.linspace(0, 10, 20)
data1
```

```python
data2 = data1 ** 2
```

```python
plt.subplots()
```

```python
my_fig, my_axes = plt.subplots()
my_axes.plot(data1, data2, alpha=0.7)
my_axes.plot(data2, data1)
plt.show()
```

```python
new_fig, new_axes = plt.subplots()
new_axes.plot(data1, data1+2, color="blue", linewidth=1.0)
new_axes.plot(data1, data1+3, color="yellow", linewidth=0.7)
new_axes.plot(data1, data1+4, color="black", linestyle="-.")
new_axes.plot(data1, data1+5, color="green", linestyle=":")
new_axes.plot(data1, data1+6, color="red", linestyle="--")
new_axes.plot(data1, data1+7, color="black", marker="o", markersize=4, markerfacecolor="red")
new_axes.plot(data1, data1+8, color="#f5b631", linestyle="--", marker="+", markersize=4)

plt.show()
```

```python
plt.scatter(data1, data2, color="red")
plt.show()
```
