```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
age_list = [10,20,30,30,40,60,65,70,80]
weight_list = [20,60,64,75,78,67,78,85,84]
```

```python
plt.plot(age_list, weight_list, color="g")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.show()
```

```python
np_age_list = np.array(age_list)
np_weight_list = np.array(weight_list) 
```

```python
plt.plot(np_age_list, np_weight_list, color="r")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age vs Weight")
plt.show()
```

```python
numpy_arr1 = np.linspace(0, 10, 20)
numpy_arr2 = numpy_arr1 ** 3
```

```python
plt.plot(numpy_arr1, numpy_arr2, color="y")
plt.show()
```

```python
plt.plot(numpy_arr1, numpy_arr2, "b--")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(numpy_arr1, numpy_arr2, "g")
plt.subplot(1,2,2)
plt.plot(numpy_arr2, numpy_arr1, "b")
plt.show()
```

```python
my_figure = plt.figure()
figure_axes = my_figure.add_axes([0.1,0.1,0.5,0.5])
figure_axes.plot(numpy_arr1, numpy_arr2, "g")
figure_axes.set_xlabel("X axis")
figure_axes.set_ylabel("Y axis")
figure_axes.set_title("Graph Title")
plt.show()
```

```python
new_fig = plt.figure()
new_axes = new_fig.add_axes([0.1,0.1,0.7,0.7])
new_axes.plot(numpy_arr1, numpy_arr1**2, label="numpy_arr1**2")
new_axes.plot(numpy_arr1, numpy_arr1**3, label="numpy_arr1**3")
new_axes.legend()
plt.show()
```
